```python
DATA_hist = np.sum(data,axis=1)
```

```python
print DATA_hist.shape
print data.shape
```

```python
fit = fit_library.gauss_fit()
fig = plt.figure()
fit(DATA_hist,'sqrt')
fit.plot(axis = fig.add_subplot(111), title = "Energy. No filter applied", 
         xlabel = "Photons", ylabel = "Hits", res = False)
plt.show()
```

```python
cond1 = (iter1 != 0)
cond1 = np.prod(cond1,axis=1,dtype='bool_')
DATA_hist_filtered = DATA_hist[cond1]
fit2 = fit_library.gauss_fit()
fig = plt.figure(figsize=(10,10))
fit2(DATA_hist_filtered,'sqrt')
fit2.plot(axis = fig.add_subplot(111), title = "Energy. Coincidence Filter", 
         xlabel = "Photons", ylabel = "Hits", res = True)
plt.show()
```
